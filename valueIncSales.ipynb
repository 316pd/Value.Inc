```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/DataViz/transaction.csv', sep=';')
```

```python
df.head()
```

```python
df['TotalCostPerTran'] = df['NumberOfItemsPurchased']*df['CostPerItem']
df['TotalSellingPricePerTran'] = df['NumberOfItemsPurchased']*df['SellingPricePerItem']
df['salesProfitPerTran'] = df['TotalSellingPricePerTran']-df['TotalCostPerTran']
df['salesPercentPerTran'] = round(df['salesProfitPerTran']/df['TotalCostPerTran'], 2)
```

```python
df['Date'] = df['Day'].astype(str)+'-'+df['Month']+'-'+df['Year'].astype(str)
df.head()
```

```python
df['ClientKeywords'][0][2:-2].replace('\'','').split(', ')
t0, t1, t2 = [], [], []
for i in range(len(df)):
  t0.append(df['ClientKeywords'][i][2:-2].replace('\'','').split(', ')[0])
  t1.append(df['ClientKeywords'][i][2:-2].replace('\'','').split(', ')[1])
  t2.append(df['ClientKeywords'][i][2:-2].replace('\'','').split(', ')[2])
```

```python
df['age'] = t0
df['orgType'] = t1
df['dealDuration'] = t2
df['ItemDescription'] = df['ItemDescription'].str.lower()
df.head()
```

```python
df2=pd.read_csv('/content/drive/MyDrive/DataViz/value_inc_seasons.csv',sep=';')
df2.head()
```

```python
df = df.merge(df2, on='Month')
```

```python
df.to_csv('ValueIncCleaned.csv',index=False)
```
